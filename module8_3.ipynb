```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df=pd.read_csv('tips.csv')
print(df)
# total_bill — общая сумма, оплаченного заказа;
# tip — размер чаевых;
# sex — пол клиента;
# smoker — является ли клиент курильщиком (в 90-е годы в ресторанах США были зоны для курящих);
# day — день недели;
# time — время (обед или ужин);
# size — количество посетителей за столом.
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=df)

```

```python
sns.kdeplot(df['total_bill'])

sns.rugplot(df['total_bill'])
```

```python
sns.pairplot(df, hue='sex', palette='Set1')
```

```python
# step 10
# Количество посетителей ресторана в зависимости от 
# дня недели с разделением по времени 
# (Lunch - Завтрак, Dinner - Обед).
sns.countplot(x='day',  hue='time', data=df)
```

```python
# step 11
# Количество клиентов ресторана в зависимости от
# количества посетителей за столом с разделением по полу
# (Female - Женщина, Male - Мужчина).
sns.countplot(x='size',  hue='sex', data=df)
```

```python
df=pd.read_csv('titanic.csv')
print(df)

# survived - Пасажир выжил (1 = Да; 0 = Нет)
# pclass - Класс каюты пассажира (1 = Первый; 2 = Второй; 3 = Третий)
# sex - Пол
# age - Возраст
# sibsp - Количество Братьев (Сестер) / Супругов на борту
# parch - Количество Родителей / Детей на борту
# fare - Пассажирский тариф
# embarked - Порт посадки (C = Cherbourg; Q = Queenstown; S = Southampton)
# class - Класс пассажира в текстовом виде ( First, Second, Third)
# who - woman_child_or_man
# adult_male - Пассажир взрослый мужчина (True = Да; False = Нет)
# deck - Палуба
# embark_town - Порт посадки в текстовом виде (Cherbourg; Queenstown; Southampton)
# alive - Пасажир выжил (Yes = Да; No = Нет)
# alone - Пассажир путешествовал один (True = Да; False = Нет)
```

```python
# step 13
# Количество пассажиров на борту Титаника в зависимости от
# класса каюты пассажира (pclass или class)
# с разделением по полу (sex: female - женщина, male - мужчина).

sns.countplot(x='class',  hue='sex', data=df)
```

```python
# step 14
# Количество пассажиров Титаника в зависимости от
# палубы размещения (deck)
# с разделением по полу (sex: female - женщина, male - мужчина).

sns.countplot(x='deck',  hue='sex', data=df)
```

```python
# step 15
# Количество выживших пассажиров Титаника в зависимости от
# палубы размещения с разделением по классу каюты.  

# sns.barplot(x='deck', y='alive', data=df)
# 

# гистограммы всех пассажиров в зависимости от
# палубы размещения с разделением по классу каюты и
# выживших пассажиров в зависимости от палубы размещения с разделением по классу каюты
sns.catplot(x="deck", hue="pclass", col="survived",data=df, kind="count")
# sns.catplot(x="deck", hue="pclass",data=df, kind="count")

```
