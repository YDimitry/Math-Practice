```python
# from math import sin, e
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from numpy.ma import sin
from numpy import e

f = lambda x: sin(x/5) * pow(e, x/10) + 5 * pow(e,-x/2) 
g = lambda x: 1 - pow(x,3) + pow(x,5) 
h = lambda x: (5 * pow(sin(pow(e,0.8*x)),2) * pow(x,2))/pow(e,pow(x,2)/2)

f_c, g_c, h_c = 'bgr'
```

$f(x)=sin(\frac{x}{5})\cdot e^{\frac{x}{10}}+5\cdot e^{-\frac{x}{2}}$

* в положительной области абсцисс функция принимает вид периодической функции с возрастающей по экспоненте аммплтудой

```python
fig, (ax1,ax2,ax3) = plt.subplots(1, 3,figsize=(15,5))


x1 = np.arange(-300,1000,0.5)
y1 = np.ma.masked_outside(f(x1),-1e25,1e25)
ax1.plot(x1,y1,f_c)
ax1.set_ylim(-3e13,1.1e14)
ax1.set_title('общее поведение функции')
ax1.grid()

x2 = np.arange(-5,50,0.1)
y2 = np.ma.masked_outside(f(x2),-20,20)
ax2.plot(x2,y2,f_c)
ax2.set_title('перегибы функции при\nмалых положительных x')
ax2.grid()

x3 = np.arange(-50,0,0.1)
y3 = f(x3) #np.ma.masked_outside(f(x3),0,20)
ax3.plot(x3,y3,f_c)
ax3.set_title('поведение функции при\nотридцательных x')
ax3.grid()

fig.suptitle(r'График функции $f(x)=sin(\frac{x}{5})\cdot e^{\frac{x}{10}}+5\cdot e^{-\frac{x}{2}}$', fontsize=18)
plt.subplots_adjust(top=0.8)
```

$g(x)=1-x^3+x^5$

* перегибы функции при $x = \pm 0,77459666924148337704


```python
fig, (ax1,ax2,ax3) = plt.subplots(1, 3,figsize=(15,5))

x1 = np.arange(-300,300,0.5)
y1 = g(x1)
ax1.plot(x1, y1,g_c)
ax1.set_title('общее поведение функции')
ax1.grid()

x2 = np.arange(-2,2,0.01)
y2 = np.ma.masked_outside(g(x2),0.8,1.2)
ax2.plot(x2, y2,g_c)
ax2.set_title('перегибы функции')
ax2.grid()

x3 = np.arange(3,6,0.01)
y3 = np.ma.masked_outside(g(x3),0,10)
ax3.plot(x3, y3,g_c)
ax3.set_title('общее поведение функции')
ax3.grid()

fig.suptitle(r'График функции $g(x)=1-x^3+x^5$', fontsize=18)
plt.subplots_adjust(top=0.8)
```

$h(x)={\large \frac{5\cdot sin^2(e^{0.8x})\cdot x^2}{e^{\frac{x^2}{2}}}}$

```python
#fig, (ax1,ax2,ax3) = plt.subplots(1, 3,figsize=(15,5))
fig = plt.figure(figsize=(15,5))

x1 = np.arange(-30,30,0.05)
y1 = h(x1)
ax1 = fig.add_subplot(131)
ax1.plot(x1, y1,h_c)
ax1.set_title('общее поведение функции')
ax1.grid()

x2 = np.arange(0,6,0.05)
y2 = h(x2)
ax2 = fig.add_subplot(132)
ax2.plot(x2, y2,h_c)
ax2.set_title('перегибы функции')
ax2.grid()

x3 = np.arange(-10,-9,0.01)
y3 = h(x3)
ax3 = fig.add_subplot(133)
ax3.plot(x3, y3,h_c)
ax3.set_title('поведение функции при\nотридцательных x')
ax3.grid()

fig.suptitle(r'График функции $h(x)=\frac{5\cdot sin^2(e^{0.8x})\cdot x^2}{e^{\frac{x^2}{2}}}$', fontsize=18)
plt.subplots_adjust(top=0.8)
```

```python

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,8))
x1 = np.arange(-3,3,0.05)
y11 = np.ma.masked_outside(f(x1),0,4)
y12 = np.ma.masked_outside(g(x1),0,4)
ax1.plot(x1,y11, f_c, label=r'f(x)') 
ax1.plot(x1,y12, g_c, label=r'$g(x)=1-x^3+x^5$')
ax1.set_title(r'первое пересечение функций $f(x)$ и $g(x)$')
ax2.set_xlabel('$x$', fontsize=18)
ax1.grid()
ax1.legend()
ax1.yaxis.set_minor_locator(AutoMinorLocator())
ax1.xaxis.set_minor_locator(AutoMinorLocator())

x2 = np.arange(-100,650,0.5)
y21 = np.ma.masked_outside(f(x2),-1e30,1e25)
y22 = np.ma.masked_outside(g(x2),-1e14,1e14)
ax2.plot(x2,y21, f_c, label=r'f(x)') 
ax2.plot(x2,y22, g_c, label=r'$g(x)=1-x^3+x^5$')
ax2.set_title(r'последующие пересечения $f(x)$ и $g(x)$')
ax2.set_xlabel('$x$', fontsize=18)
ax2.set_ylim(-3e13,1.1e14)
ax2.grid()
ax2.legend()
ax2.yaxis.set_minor_locator(AutoMinorLocator())
ax2.xaxis.set_minor_locator(AutoMinorLocator())

fig.suptitle(r'Пересечения функций $f(x)$ и $g(x)$', fontsize=18)
```
