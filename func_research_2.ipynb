```python
# from math import sin, e
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from numpy.ma import sin
from numpy import e

f = lambda x: sin(x/5) * pow(e, x/10) + 5 * pow(e,-x/2) 
# f = lambda x: sin(x/5)
g = lambda x: 1 - pow(x,3) + pow(x,5) 
h = lambda x: (5 * pow(sin(pow(e,0.8*x)),2) * pow(x,2))/pow(e,pow(x,2)/2)

y1_c, y2_c, y3_c, y4_c, y5_c, y6_c = 'rbgcmy'
```

```python
x = np.arange(-100,2000,0.1)
y1 = f(x) #np.ma.masked_outside(f(x),-200,200)
y2 = np.ma.masked_outside(g(x),-200,200)
y3 = np.ma.masked_outside(h(x),-200,200)

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y1, y1_c, label=r'f(x)') 
# ax.plot(x,y2, y2_c, label=r'$g(x)=1-x^3+x^5$')
# ax.plot(x,y3, y3_c, label=r'h(x)')
ax.set_title(r'График функции $f(x)=\frac{x^\beta+\alpha^\beta}{x^\beta}$ при различных $\alpha$ и $\beta$', fontsize=18)
ax.set_xlabel('$x$', fontsize=18)
ax.set_ylabel('$f(x)$', fontsize=18)
ax.set_ylim(-5e50,5e50)
ax.grid()
ax.legend()
ax.yaxis.set_minor_locator(AutoMinorLocator())

plt.show()
```
